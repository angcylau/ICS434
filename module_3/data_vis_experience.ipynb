#### 1. Reproducing a plot
1. Use `numpy`'s `random` module to generate a random list of size 1000 of the following three categories: "CS", "Engineering", "Math" and "Physics"
  1.1 Use a seed of 42
2. Store the data in an array called my `students_affiliations`
3. use `matplotlib` to plot the data in `students_affiliations` to reproduce the following graph
 3.1 The plot uses "#FF5733" as color with an alpha (transparency) of 0.55.

* Your graph should look exactly like the one below

![](https://www.dropbox.com/s/ffmu62es6d7evh5/experience_fig.png?dl=1)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Using the `random` module from `numpy`, the seed is set to 42 so that the same random data is reproducible. <br><br>
`rand_discipline` is an array that uses `np.random.choice` to randomly select one of the four disciplines (CS, Engineering, Math, or Physics) 1000 times, with replacement. <br><br>
`student_affiliations` is a Pandas Series that takes the array of randomly selected disciplines and generates the counts of each of the four disciplines.

```python
np.random.seed(42)
rand_discipline = np.random.choice(['CS', 'Engineering', 'Math', 'Physics'], size=1_000, replace=True)
student_affiliations = pd.Series(rand_discipline).value_counts()
student_affiliations.name = 'Counts'

student_affiliations
```

Using `matplotlib.pyplot` generate a bar plot, where the x-values are the disciplines and the y-values are the counts. The `color` of the bars are set with a hex value and the `alpha` is the transparency of the color, set with a numerical value of 0-1.

```python
plt.figure(figsize=(12,8))
x = student_affiliations.index
y = student_affiliations.values
plt.bar(x, y,  color='#FF5733', alpha=0.55)
plt.xlabel('Disciplines', fontsize=20)
plt.ylabel('Counts', fontsize=20)
plt.title('Distribution of Students in the Program', fontsize=22, color='#389437')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

#### 2. Generating a scatter plot
* Generate random data to produce a scatter plot
  * You can use any function to generate the data
  * Instead of using circles as symbols, you should use squares.

Set the random seed to 42 and create random integer values for the x and y axes using `np.random.randint`. Here, `weight_kg` selects 100 random integers from 35 to 160 and `height_cm` selects 100 random integers from 130 to 215. <br><br>
`plt.scatter` creates a scatter plot using `weight_kg` and `height_cm` as its x and y axes, and `marker='s'` sets the markers as squares.

```python
np.random.seed(42)
weight_kg = np.random.randint(35, 160, 100)
height_cm = np.random.randint(130, 215, 100)

plt.figure(figsize=(12,8))
plt.scatter(weight_kg, height_cm, marker='s')

plt.xlabel('Weight (kg)', fontsize=14)
plt.ylabel('Height (cm)', fontsize=14)
plt.title('Representation of Height and Weight', fontsize=16)
```
